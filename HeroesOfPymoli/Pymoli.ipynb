```python
import pandas as pd
filepath = "../../UT-TOR-DATA-PT-09-2019-U-C/Unit 4 - Pandas/Homework/Instructions/HeroesOfPymoli/Resources/purchase_data.csv"
```

```python
df = pd.read_csv(filepath)
df1 = pd.DataFrame(df)
df1.head()
```

```python
# Total Number of Players
user_counts = df["SN"].value_counts()
unique_users = pd.DataFrame(user_counts)
n_of_users = unique_users.count() 
n_of_users
```

```python
# Number of Unique Items
purchase_counts = df["Item ID"].value_counts()
unique_items = purchase_counts.count()
unique_items
```

```python
# Average Price
average_price = df["Price"].sum()/ df["Price"].count()
average_price
```

```python
# Total number of purchases
number_of_purchases = df["Purchase ID"].count()
number_of_purchases
```

```python
# Total Revenue
total_revenue = df["Price"].sum()
total_revenue
```

# Gender Demographics

```python
# Percentage and Count of Male Players

count_of_male = df.groupby(["SN","Gender"])
count_of_male1 = pd.DataFrame(count_of_male.size())
count_of_male2 = pd.DataFrame(count_of_male1.groupby(["Gender"]).count())
gender_df = count_of_male2.rename(columns = {0:"Size"})
gender_df
```

```python
#percentage and count of genders
gender_df["% of population"] = gender_df["Size"]/gender_df["Size"].sum() * 100
gender_df
```

# Purchasing Analysis by Gender

```python
# Purchase Count
purchase_count = df.groupby(["Gender"])["Purchase ID"].count()
purchase_count_df = pd.DataFrame(purchase_count)
gender_pc_df = purchase_count_df.rename(columns = {"Purchase ID":"Purchase Count"})
gender_pc_df
```

```python
# Average Purchase Price
avp = df.groupby(["Gender"])["Price"].mean()
avp_df = pd.DataFrame(avp)
avp_df = avp_df.rename(columns = {"Price":"Average Purchase Price"})
avp_df
```

```python
# Total Purchase Value
tpv = df.groupby(["Gender"])["Price"].sum()
tpv_df = pd.DataFrame(tpv)
tpv_df = tpv_df.rename(columns = {"Price":"Total Purchase Value"})
tpv_df
```

```python
# Average Purchase Total per Person by Gender
merge_test = pd.merge(gender_pc_df,avp_df, on = "Gender")
merge_test2 = pd.merge(merge_test,tpv_df, on= "Gender")
merge_test2["Average Purchase Total per Person by Gender"]= merge_test2["Total Purchase Value"]/gender_df["Size"]
merge_test2
```

# Age Demographics

```python
## Age Demographics
bins = [6, 9, 14, 19, 24, 29, 34, 39, 44, 47]

group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40-44",">44"]

df["Age Group"] = pd.cut(df["Age"], bins, labels=group_names)
```

```python
# Purchase Count 
AD_purchasecount = df.groupby(["Age Group"])["Purchase ID"].count()
AD_purchasecount1 = pd.DataFrame(AD_purchasecount)
AD_purchasecount2 = AD_purchasecount1.rename(columns = {"Purchase ID":"Purchase Count"})
AD_purchasecount2
```

```python
# Average Purchase Price 
AD_app = df.groupby("Age Group")["Price"].mean()
AD_app1 = pd.DataFrame(AD_app)
AD_app2 = AD_app1.rename(columns= {"Price":"Average Purchase Price"})
AD_app2
```

```python
# Total Purchase Value 
AD_tpv = df.groupby("Age Group")["Price"].sum()
AD_tpv1 = pd.DataFrame(AD_tpv)
AD_tpv2 = AD_tpv1.rename(columns = {"Price":"Total Purchase Value"})
AD_tpv2
```

```python
# Average Purchase Total per Person by Age Group
merge = pd.merge(AD_purchasecount2, AD_app2, on="Age Group")
merge2 = pd.merge(merge,AD_tpv2, on ="Age Group")
merge2
```

# Top 5 Spenders 

```python
#top 5 spenders by total purchase value
topspenders = df.groupby("SN")["Price"].sum()
topspenders_df = topspenders.sort_values(ascending=False)
#topspenders average


topspenders_df_total_value = pd.DataFrame(topspenders_df.head())
topspenders_df_total_value.rename(columns = {"Price":"Total Purchase Value"})


#dataframe for average purchase price per SN

#dataframe for COUNT of items per SN
#merge all three
```

```python
# Average Purchase Price
topspenders_avg = df.groupby("SN")["Price"].mean()
topspenders_avg_df = topspenders_avg.sort_values(ascending=False)
topspenders_avg_df_1 = pd.DataFrame(topspenders_avg_df)
topspenders_avg_df2 = topspenders_avg_df_1.rename(columns = {"Price":"Average Purchase Price"})
topspenders_avg_df2.head()

```

```python
topspenders_count = df.groupby("SN").count()

topspenders_count_df = pd.DataFrame(topspenders_count["Purchase ID"])
topspenders_count_df1 = topspenders_count_df.rename(columns ={"Purchase ID":"Purchase Count"})
topspenders_count_df2 = topspenders_count_df1.sort_values("Purchase Count",ascending=False)
topspenders_count_df2.head()
```

```python
merge_table = pd.merge(topspenders_df_total_value, topspenders_avg_df2, on="SN")
merge_table2 = pd.merge(merge_table,topspenders_count_df2, on="SN")
merge_table2
```

# Most Popular Items

```python
#Purchase Count
popularitems = df.groupby("Item ID")["Purchase ID"].count()
popularitems_df = popularitems.sort_values(ascending=False)
popularitems_df1 = pd.DataFrame(popularitems_df)
popularitems_df2 = popularitems_df1.rename(columns = {"Purchase ID":"Purchase Count"})
popularitems_df2.head()
```

```python
# Total Purchase Value 
totalpurchasevalue = df.groupby("Item ID")["Price"].sum()
totalpurchasevalue_df = totalpurchasevalue.sort_values(ascending = False)
tpv_df = pd.DataFrame(totalpurchasevalue_df)
tpv_df1 = tpv_df.rename(columns = {"Price":"Total Purchase Value"})
tpv_df1.head()
```

```python
#Item Name 
itemname = df1.iloc[:,[4,5]]
dfitemname = pd.DataFrame(itemname)
dfitemname2 = dfitemname.drop_duplicates()
dfitemname2.head()
```

```python
#Price
popularitem_price = df1.iloc[:,[4,6]]
popularitem_price1 = popularitem_price.drop_duplicates()
```

```python
merge_popular = pd.merge(popularitems_df2,tpv_df1,on="Item ID")
merge_popular1 = pd.merge(merge_popular,dfitemname2,on="Item ID")
merge_popular2 = pd.merge(merge_popular1,popularitem_price1,on="Item ID")
merge_popular2.head()
```

# Most Profitable Items

```python
# Total Purchase Value 
profitableitems = df.groupby(["Item ID"])["Price"].sum()
profitableitems_df = profitableitems.sort_values(ascending=False)
profitableitems_df1 = pd.DataFrame(profitableitems_df)
mpi_tpv_df = profitableitems_df1.rename(columns = {"Price":"Total Purchase Value"})
mpi_tpv_df.head()
```

```python
# Item Name
MPI_itemname = df1.iloc[:,[4,5]]
MPI_itemname1 = MPI_itemname.drop_duplicates()
MPI_itemname1.head()
```

```python
# Purchase Count
MPI_purchasecount = pd.DataFrame(df.groupby(["Item ID"])["Purchase ID"].count())
mpi_pc = MPI_purchasecount.rename(columns = {"Purchase ID":"Purchase Count"})
mpi_pc.head()
```

```python
# Price Column

price = df1.iloc[:,[4,6]]
dfprice = pd.DataFrame(price)
dfprice1 = dfprice.sort_values(by="Item ID", ascending=True)
dfprice2 = dfprice.drop_duplicates()
dfprice2.head()
```

```python
# Merging
merge_MPI = pd.merge(mpi_tpv_df,MPI_itemname1,on="Item ID")
merge_MPI_1 = pd.merge(merge_MPI, mpi_pc, on = "Item ID")
merge_MPI_2 = pd.merge(merge_MPI_1, dfprice2, on = "Item ID")
merge_MPI_2.head()
```
